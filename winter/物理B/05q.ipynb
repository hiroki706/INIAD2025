# 問題
>1辺が$15[cm]$の正方形の頂点に大きさ$1.2×10-5[C]$の
正の電荷と負の電荷が順番においてある．クーロン定数
を$k=9.0×109[N･m2/C2]$として，各電荷に働く力$F[N]$
を求めよ． 

いいですね！この問題は **電気力（クーロン力）とベクトルの合成** を扱う典型的な問題です。
丁寧に説明していきます。

---

## 🧩 問題の整理

1辺が
$$
a = 15\ \text{cm} = 0.15\ \text{m}
$$
の正方形があります。

各頂点に電荷を「正・負・正・負」と交互に置きます。つまり：

| 頂点 | 電荷 |
| -- | -- |
| A  | $+q$ |
| B  | $−q$ |
| C  | $+q$ |
| D  | $−q$ |

とします（順番に置かれている、という記述からこう解釈できます）。

電荷の大きさは
$$
q = 1.2\times 10^{-5}\ \text{C}
$$
です。

クーロン定数は
$$
k = 9.0\times 10^9 [\text{N} \cdotp m^2/\text{C}^2]
$$

---

## ⚡ 各電荷に働く力

1つの電荷（例えばAの+q）に働く力は、他の3つの電荷からのクーロン力の **ベクトル和** です。

### 1️⃣ 距離の確認

* 隣り合う頂点間の距離：( a = 0.15\text{ m} )
* 対角線の距離：( a\sqrt{2} = 0.15\sqrt{2} \text{ m} )

---

## 💥 A(＋q) に働く力

他の電荷との関係を見ます。

| 相手 | 位置関係  | 電荷 | 方向   | 作用        |
| -- | ----- | -- | ---- | --------- |
| B  | 隣（右隣） | $−q$ | 引き合う | 右向き       |
| D  | 隣（下隣） | $−q$ | 引き合う | 下向き       |
| C  | 対角線上  | $+q$ | 反発する | 対角線（左下）方向 |

---

### 各力の大きさ

#### (1) 隣の電荷（BまたはD）との力

$$
F_1 = k\frac{q^2}{a^2}
$$

#### (2) 対角線の電荷（C）との力

$$
F_2 = k\frac{q^2}{(a\sqrt{2})^2} = \frac{kq^2}{2a^2}
$$

---

### ベクトル合成

Bから右方向へ ( $F_1$ )、
Dから下方向へ ( $F_1$ )、
Cから左下方向へ ( $F_2$ )。

Cからの力を成分に分けると：

* $x成分（左向き）＝ (-\frac{F_2}{\sqrt{2}})$
* $y成分（下向き）＝ (-\frac{F_2}{\sqrt{2}})$

よって、Aに働く合力の成分は：

$$
\begin{cases}
F_x = +F_1 - \frac{F_2}{\sqrt{2}}\
F_y = -F_1 - \frac{F_2}{\sqrt{2}}
\end{cases}
$$

---

## 🧮 数値計算

### まず ( $F_1$ ) と ( $F_2$ ) を求めます。

$$
F_1 = 9.0\times10^9 \times \frac{(1.2\times10^{-5})^2}{(0.15)^2}
$$

1. $( (1.2\times10^{-5})^2 = 1.44\times10^{-10} )$
2. 分母 $( (0.15)^2 = 0.0225 )$
3. よって
   $$
   F_1 = 9.0\times10^9 \times \frac{1.44\times10^{-10}}{0.0225}
   = 9.0\times10^9 \times 6.4\times10^{-9}
   = 57.6\ \text{N}
   $$

$$
F_2 = \frac{F_1}{2} = 28.8\ \text{N}
$$

---

### 成分を代入

$$
\begin{cases}
F_x = 57.6 - \frac{28.8}{\sqrt{2}} = 57.6 - 20.4 = 37.2\ \text{N}\
F_y = -57.6 - 20.4 = -78.0\ \text{N}
\end{cases}
$$

---

### 合力の大きさ

$$
F = \sqrt{F_x^2 + F_y^2}
= \sqrt{(37.2)^2 + (78.0)^2}
= \sqrt{1384 + 6084}
= \sqrt{7468}
\approx 86.4\ \text{N}
$$

---

## ✅ 答え

各電荷に働く力の大きさ：

$$
\boxed{F = 8.6\times 10^{1}\ \text{N}}
$$

方向は：

* +q には対角線の方向に向かう（正から負へ引かれる）向き。

---

## 🧠 ポイントまとめ

* クーロン力はベクトルなので、方向を必ず分解して合成！
* 同符号 → 反発、異符号 → 引力。
* 正方形配置の問題は、対称性と成分分解を使うと整理しやすい。
