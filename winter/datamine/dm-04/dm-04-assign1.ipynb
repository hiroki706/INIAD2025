```python
import pandas as pd
import folium
```

```python
csv_in = './crime2006_2016-utf8.csv'

pd.options.display.max_rows = 999
pd.options.display.max_columns = 999
```

```python
df = pd.read_csv(csv_in, sep=',', skiprows=0, header=0)
df = df.where(df["年"] == 2016)
print(df.shape)
print(df.info())
display(df.head())
#display(df)
```

```python
df_pos = pd.read_csv("./h3010pref_pos_utf8.csv", comment="#", sep=",")
print(df_pos.shape)
print(df_pos.info())
display(df_pos.head())
```

```python
df_merged = pd.merge(df, df_pos, left_on='都道府県', right_on='name', how='inner')
print(df_merged.info())
print(df_merged.shape)
display(df_merged.head())
```

```python
df_heat = df_merged[["lat", "lon", "検挙人数", ]]
print(df_heat.shape)
display(df_heat.head())
```

```python
from folium.plugins import HeatMap

m = folium.Map(location=[35, 140], zoom_start=4)
m.add_child(HeatMap(df_heat, radius=12))
display(m)
```
