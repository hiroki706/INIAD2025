```python
#Cell_2.
csv_in = 'gld_price_data.csv'
df = pd.read_csv(csv_in, sep=',', skiprows=0, header=0)
display(df.head())
df['Date'] = pd.to_datetime(df['Date'], format="%m/%d/%Y")
```

```python
#Cell_3.
print(df.info())
```

```python
#Cell_4.
df_gld = df['GLD']
df_gld.index = df['Date']

df_monthly = df_gld.resample('M').mean() 
print(df_monthly.dtype)
display(df_monthly.head())
```
