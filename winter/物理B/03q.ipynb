# 問題
>圧縮比$ \frac{V_L}{V_H}=\frac{1}{8} $のオットーサイクルに$ 27[℃] $の理想気体$ 1 [mol]$を入れた。比熱比を$ γ=\frac{5}{3} $として熱効率$η [\%]$を求めよ。

答え: $ 75[\%] $

# 回答
## 1. オットーサイクルの基本

オットーサイクルは

1. 等容加熱
2. 断熱膨張
3. 等容冷却
4. 断熱圧縮

の4つの過程から成り立ちます。
熱効率 ($\eta$) は、投入熱量 ($Q_H$) と出力仕事 ($W$) の比から

$$
\eta = \frac{W}{Q_H} = 1 - \frac{Q_L}{Q_H}
$$

で表されます。

---

## 2. 断熱変化と比熱比

断熱変化においては「ポアソンの法則」

$$
PV^\gamma = \text{一定}
$$

が成り立ちます。
ここで比熱比 (\gamma) は

$$
\gamma = \frac{C_p}{C_v}
$$

です。

この関係を使うことで、断熱圧縮比 ($r = \frac{V_H}{V_L}$) に依存して温度比を表すことができます。

---

## 3. 熱効率の導出

詳細な計算を整理すると、オットーサイクルの熱効率は

$$
\eta = 1 - \frac{1}{r^{\gamma-1}}
$$

という単純な式になります。
この式の特徴は、カルノーサイクルのように温度でなく、**圧縮比 ($r$) と比熱比 ($\gamma$) のみに依存する**点です。

---

## 4. 与えられた条件の適用

* 圧縮比 ($r = 8$)
* 比熱比 ($\gamma = 5/3$)

したがって、

$$
\eta = 1 - \frac{1}{8^{\tfrac{5}{3}-1}}
= 1 - \frac{1}{8^{2/3}}
= 1 - \frac{1}{4}
= 0.75
$$

すなわち **熱効率は 75%** となります。

---

## まとめ

* 資料にあるように、オットーサイクルの効率は ($r$) と ($\gamma$) のみに依存する。
* 与えられた値を代入すると ($\eta = 75[\%]$)。
* 初期温度やモル数は効率式には影響を与えないため、最終的な答えは **75%** です。
