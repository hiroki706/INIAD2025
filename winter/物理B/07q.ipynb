# 問題
>電荷$79e$の金の原子核に電荷$2e$，質量$4u$のアルファ粒子を無限遠から初速度$2.0×107[m/s]$で正面衝突させた．素電荷を$e=1.6×10-19[C]$，$u=1.66×10-27[kg]$，真空の誘電率を$ε0=8.85×10-12[C･m/V]$として，アルファ粒子が最も近づく距離$r0[fm]$を求めよ．

解き方は「運動エネルギー＝反発の静電位エネルギー（最近接時）」です。真正面衝突なので、最近接距離 ($r_0$) で速度は $0$ になり、初期の運動エネルギーがすべて静電位エネルギーに変わります。

$[
\frac12 m v^2 ;=; \frac{1}{4\pi\varepsilon_0},\frac{(79e),(2e)}{r_0}
]$

よって

$[
r_0 ;=; \frac{1}{4\pi\varepsilon_0},\frac{(79 \cdot 2)e^2}{\tfrac12 m v^2}
= \frac{1}{2\pi\varepsilon_0},\frac{(79 \cdot 2)e^2}{m v^2}
]$

数値代入
$$
(m=4u=4 \times 1.66 \times 10^{-27}=6.64 \times 10^{-27},\text{kg})\\
(v=2.0 \times 10^{7}, \text{m/s})\\
(e=1.6 \times 10^{-19}, \text{C})\\
(\varepsilon_0=8.85 \times 10^{-12},\text{C /cdotp m/V})\\
$$


* 初期運動エネルギー
$$
  (\displaystyle \frac12 m v^2
  = \frac12 \times 6.64\times10^{-27} \times (2.0\times10^{7})^2
  = 1.328\times10^{-12}\ \text{J})
$$
* 分子（クーロン定数 $(k=1/(4\pi\varepsilon_0))$ を使うと）
$$
  (\displaystyle k,(79 \cdot 2),e^2
  = \frac{1}{4\pi\varepsilon_0}\times 158 \times (1.6\times10^{-19})^2
  \approx 2.739\times10^{-14}\ \text{J /cdotp m})
$$
したがって
$
[
r_0 ;=; \frac{2.739\times10^{-14}}{1.328\times10^{-12}}
;=; 2.74\times10^{-14}\ \text{m}
;=; 27.4\ \text{fm}.
]
$
## 答え
$
[
\boxed{r_0 \approx 27\ \text{fm}}
]$
（有効数字2桁で表記）。
※ これは金原子核の半径（数 fm）より十分大きく、古典的なクーロン反発の近似が妥当なことも確認できます。
