```python
import openai
import re
import os
from openai import OpenAI
from openai.types.chat import ChatCompletionMessageParam
from mylocalenv import OPENAI_API_KEY

oai = OpenAI(
    api_key = OPENAI_API_KEY,
    base_url = "https://api.openai.iniad.org/api/v1"
)
```

```python
with open('./toyo_students.csv', 'r', encoding='utf-8') as file:
    file_content = file.read()
```

```python
#Cell_3.
from typing import List
messages: List[ChatCompletionMessageParam] =[
       {"role": "system", "content": "あなたはデータサイエンティストです。"},
       {"role": "user", "content": "いまアップロードしたデータ'toyo-sutudent.csv'について解説してください。"},
       {"role": "user", "content": file_content}
]

resp = oai.chat.completions.create(
    model='gpt-4o-mini',
    messages=messages,
)

print(resp.choices[0].message.content)
```

```python
#Cell_4.
messages.append({"role": "assistant", "content": resp.choices[0].message.content})
messages.append({"role": "user", "content": "東洋大学の男性学生数、女性学生数の合計（全キャンパスの合計）を求めて、どちらが多いかを報告してください。"})

resp = oai.chat.completions.create(
    model='gpt-4o-mini',
    messages=messages,
)

print(resp.choices[0].message.content)
```
