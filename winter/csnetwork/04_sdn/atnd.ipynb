```python
import ipaddress

networks = [
    ipaddress.ip_network("0.0.0.0/0"),
    ipaddress.ip_network("192.168.0.0/22"),
    ipaddress.ip_network("192.168.4.0/22"),
    ipaddress.ip_network("192.168.8.0/23"),
    ipaddress.ip_network("192.168.10.0/24"),
]
dist = ipaddress.ip_address("192.168.15.132")

# ネットワークの範囲内でプレフィックスが最大限なものを選ぶ
matchip = max([nw for nw in networks if dist in nw])
matchip
```

```python
import rustworkx as nx
G = nx.PyGraph()

a = G.add_node("a")
b = G.add_node("b")
c = G.add_node("c")
d = G.add_node("d")
e = G.add_node("e")

G.add_edges_from([
    (a, b, 1),
    (a, c, 3),
    (b, d, 1),
    (b, e, 4),
    (c, d, 3),
    (d, e, 8)
])
print(G.nodes())
res = nx.dijkstra_shortest_paths(G, source=a, target=e, weight_fn=float)
aaa = list(G.nodes())
for i in res.values():
    for j in i:
        print(aaa[j])
```
