```python
import pandas as pd
import numpy as np
import pyfpgrowth

df=pd.read_csv('market.csv')
```

```python
print(df.shape)
print(df.info())
display(df.head())
```

```python
top_sc = df['小カテゴリ名'].value_counts()
print(top_sc.size)
print(top_sc.head())
```

```python
id2sc = sorted(list(set(df['小カテゴリ名'])))
sc2id = {}
for i in range(len(id2sc)):
    sc2id[id2sc[i]] = i
```

```python
print(sc2id)
```

```python
df['小カテゴリ名_ID'] = df['小カテゴリ名'].map(lambda x: sc2id[x])
display(df.head())
```

```python
invoices = []
for r in df.groupby('レシートNo'):
    ##print(len(r))  # debug
    #print(r[0])  # debug
    ##print(r[1])  # debug
    #break  # debug
    s1 = set(r[1]['小カテゴリ名_ID'])
    invoices.append(list(s1))
    print(list(s1))
print(len(invoices))
```

```python
%time patterns = pyfpgrowth.find_frequent_patterns(invoices, 40)
```

```python
%time rules = pyfpgrowth.generate_association_rules(patterns, 0.5)
```

```python
print(rules)
```

```python
results = []
for x in rules:
    ret = [x, rules[x][0], rules[x][1]]
    results.append(ret)
df_res = pd.DataFrame(results)
df_res.columns = ['LHS', 'RHS', 'Conf']
```

```python
display(df_res.sort_values(by='Conf', ascending=False))
```

```python
print(id2sc[44])
print(id2sc[296])
print(id2sc[401])
```

```python
n_all = len(invoices)
lift = []
for i in range(df_res.shape[0]):
    rhs = df_res.at[i, 'RHS']
    conf = df_res.at[i, 'Conf']
    n_rhs = 0
    for items in invoices:
        if set(items) >= set(rhs):
            n_rhs += 1
    lift1 = conf / (n_rhs / n_all)
    lift.append(lift1)
    
df_res['Lift'] = lift
```

```python
display(df_res.sort_values(by='Conf', ascending=False))
```
