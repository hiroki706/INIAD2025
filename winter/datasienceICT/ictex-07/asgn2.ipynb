```python
#Cell_1.
import openai
import re
import os
from openai import OpenAI
from openai.types.chat import ChatCompletionMessageParam
from mylocalenv import OPENAI_API_KEY

oai = OpenAI(
    api_key = OPENAI_API_KEY,
    base_url = "https://api.openai.iniad.org/api/v1"
)
```

```python
#Cell_2.
with open('./used_motorcycle_en.csv', 'r', encoding='utf-8') as file:
    file_content = file.read()
```

```python
#Cell_3.
from typing import List
messages: List[ChatCompletionMessageParam] =[
       {"role": "system", "content": "あなたはデータサイエンティストです。"},
       {"role": "user", "content": "いまアップロードしたデータ'used_motorcycle.csv'について解説してください。"},
       {"role": "user", "content": file_content}
]

resp = oai.chat.completions.create(
    model='gpt-4o-mini',
    messages=messages,
)

print(resp.choices[0].message.content)
```

```python
#Cell_4.
messages.append({"role": "assistant", "content": resp.choices[0].message.content})
messages.append({"role": "user", "content": "このデータに基づいて中古バイク価格を説明するモデルを作りたいですが、上記データに含まれていないような潜在的な要因を予測確率順に出力してほしいです"})

resp = oai.chat.completions.create(
    model='gpt-4o-mini',
    messages=messages,
)

print(resp.choices[0].message.content)
```

```python
#Cell_4.
messages.append({"role": "assistant", "content": resp.choices[0].message.content})
messages.append({"role": "user", "content": "その要因をパス図にしてほしいです。"})

resp = oai.chat.completions.create(
    model='gpt-4o-mini',
    messages=messages,
)

print(resp.choices[0].message.content)
```
