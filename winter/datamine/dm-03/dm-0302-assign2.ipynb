## Create simple plot with Pandas  

### Import libraries

```python
import pandas as pd
```

### Parameters

```python
csv_in1 = '2045_2017-utf8.csv'
```

### Read CSV file

```python
df1 = pd.read_csv(csv_in1, sep=',', skiprows=9, header=0)
df1['Date']=pd.to_datetime(df1['Date'], format="%Y-%m-%d")
print(df1.shape)
print(df1.info())
display(df1.head())
```

### Add 25days moving average line  

```python
df1['25d_mov_ave'] = df1['AdjClose'].rolling(window=25).mean()
display(df1.head(30))
```

### Line plot  

```python
df1.plot.line('Date', ['AdjClose', '25d_mov_ave'])
```

### Scatter plot  

```python
df1.plot.scatter('High', 'Low')
```

```python
df2 = pd.read_csv("./1393_2017-utf8.csv", skiprows=9, sep=",")
df2["Date"] = pd.to_datetime(df2["Date"])
print(df2.shape)
print(df2.info())
display(df2.head())
```

```python
df12 = pd.merge(df1.drop(columns=["25d_mov_ave"]), df2, on="Date", how="inner")
print(df12.shape)
print(df12.info())
display(df12.head())
```

```python
df12.plot.scatter("AdjClose_x", "AdjClose_y")
```
