```python
import rustworkx as nx
```

```python
G = nx.PyGraph()

u = G.add_node("u")
a = G.add_node("a")
b = G.add_node("b")
c = G.add_node("c")
d = G.add_node("d")
e = G.add_node("e")

G.add_edges_from([
    (u, a, 5),
    (u, c, 9),
    (a, c, 3),
    (a, b, 6),
    (c, d, 4),
    (b, d, 3),
    (b, e, 13),
    (d, e, 5)
])
print(G.nodes())
from rustworkx.visualization import mpl_draw
mpl_draw(G, with_labels=True, edge_labels=str, labels=str)
```

u,a,c,b,d,e

```python
res = nx.dijkstra_shortest_paths(G, source=u, target=e, weight_fn=float)
aaa = list(G.nodes())
for i in res.values():
    for j in i:
        print(aaa[j])
```
